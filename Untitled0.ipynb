```python
# caiani_minimal_fullscale.py
# Caiani-like minimal ABM-SFC skeleton (baseline params reflected)
# Run with: python caiani_minimal_fullscale.py

import numpy as np
import random
import matplotlib.pyplot as plt

# reproducibility
SEED = 12345
random.seed(SEED)
np.random.seed(SEED)

# --- Baseline params (Caiani - from provided table) ---
params = {
    # scale
    'N_households': 8000,     # sizeΦH
    'N_firms_c': 100,         # sizeΦC (consumption firms)
    'N_firms_k': 20,          # sizeΦK (capital firms) -- not fully used in minimal
    'N_banks': 10,            # sizeΦB
    'Nc0': int(0.5 * 8000),  # Initial employees in consumption firms (approximation)
    'Nk0': 0,                # Initial employees in capital firms (not fully used)

    # time
    'T': 200,                 # periods to simulate

    # macro
    'gSS': 0.0075,            # steady growth

    # tech / production
    'muN': 2.0,               # K sector labor productivity (approx)
    'mu_k': 1.0,
    'lk': 6.4,

    # market matching/contact
    'chi_c': 5,
    'chi_k': 5,
    'chi_d': 3,
    'chi_l': 3,
    'chi_n': 10,


    # selection intensities
    'eps_d': 2.00687,
    'eps_l': 2.00687,
    'eps_c': 1.50515,
    'eps_k': 1.50515,

    # inventories, expectations
    'nu': 0.1,
    'lambda_adapt': 0.25,
    'theta_turnover': 0.05,

    # price/wage noise (folded normal params simplified)
    'mu_FN': 1.0,
    'sigma2_FN': 0.0094,

    # taxation
    'tau_pi': 0.18,
    'tau_i': 0.18,

    # lending / capital
    'eta': 20,                # loan term
    'kappa': 20,              # capital depreciation years
    'r_target': 0.04345,      # target profit rate
    'u_target': 0.8,            # target capacity utilization

    # investment weights
    'gamma1': 0.01,
    'gamma2': 0.02,

    # buffers, payout rates
    'sigma_buffer': 1.0,
    'rho_c': 0.9,
    'rho_k': 0.9,
    'rho_b': 0.6,

    # asset / bond
    'i_b0': 0.0025,
    'p_b': 1.0,

    # bank regulation targets
    'CRT0': 0.17996,
    'LRT0': 0.258026342,

    # bank risk aversion
    's_c': 3.92245,
    's_k': 21.51335,

    # bankruptcy haircut
    'haircut': 0.5,

    # wage / unemployment
    'wn0': 5.0,
    'omega_dole': 0.4,
    'v_unemp': 0.08,

    # consumption propensities
    'alpha1': 0.38581,   # income propensity
    'alpha2': 0.25,      # wealth propensity

    # initial stocks (rough baseline, scaled)
    'h0_deposit': 50.0,   # initial deposit per household
    'f0_capital': 500.0,  # initial capital per firm
    'f0_debt': 50.0,      # initial debt per firm
    'b0_capital': 50000.0, # initial capital per bank (aggregate large)

    # Missing parameter
    'markup_c0': 0.318857, # Added initial markup for consumption firms
    "xi_match": 0.5,        # マッチング効率 ξ
    "alpha_match": 0.6,     # 求人弾力性 α
    "eta_w": 0.1,           # 賃金調整速度
    "w_ref": 5.0,           # Reference wage for firms (using wn0)
}

# --- Agent classes ---

class Household:
    def __init__(self, hid, deposit, params):
        self.id = hid
        self.deposit = float(deposit)
        self.income = 0.0
        self.wealth = float(deposit)
        self.consumption = 0.0
        self.params = params
        # --- 追加: 雇用関連 ---
        self.employed = False      # 雇用フラグ（初期は False、後で割当）
        self.firm_id = None        # 所属企業 id（雇われていなければ None）
        self.wage = 0.0

    def update_employment(self, employed, firm_id=None, wage=0.0):
        self.employed = bool(employed)
        self.firm_id = firm_id if employed else None
        self.wage = float(wage) if employed else 0.0

    def receive_income(self, wage_income):
        # wage_income は賃金 or 失業給付など（呼び出し側で決定）
        self.income = float(wage_income)
        self.deposit += self.income
        self.wealth = self.deposit  # 簡略: wealth == deposit

    def decide_consumption(self):
        # C = alpha1 * income + alpha2 * wealth
        a1 = self.params['alpha1']
        a2 = self.params['alpha2']
        self.consumption = a1 * self.income + a2 * self.wealth
        # limit consumption to available deposit
        if self.consumption > self.deposit:
            self.consumption = self.deposit
        return self.consumption

    def pay(self, amount):
        paid = min(amount, self.deposit)
        self.deposit -= paid
        self.wealth = self.deposit
        return paid


class FirmC:
    def __init__(self, fid, capital, debt, params):
        self.id = fid
        self.capital = float(capital)
        self.debt = float(debt)
        self.inventory = 0.0
        self.price = 1.0 + params['markup_c0']  # initialize with markup
        self.wage = params['w_ref']  # initial wage paid to its workers
        self.productivity = 1.0   # baseline productivity (could vary)

        # employees をオブジェクトのリストで管理
        self.employees = []       # list of Household objects
        self.output = 0.0
        self.sales = 0.0
        self.profit = 0.0
        self.params = params
        # expectation memory for demand
        self.past_demand = []
        # target inventory
        self.inv_target_rate = params['nu']
        # vacancies（求人数）
        self.vacancies = 0

    def num_employees(self):
        return len(self.employees)

    def produce(self, labor_input=None):
        # if labor_input None -> use current employees
        L = self.num_employees() if labor_input is None else labor_input
        A = self.productivity
        self.output = A * L
        self.inventory += self.output

    def set_price(self, aggregate_avg_profit):
        kappa_p = 0.02
        if self.sales > 0:
            pi_i = (self.price * self.sales - self.wage * self.num_employees() - 0.05 * self.debt) / (self.price * self.sales)
        else:
            pi_i = 0.0
        gap = pi_i - aggregate_avg_profit
        self.price = max(0.01, self.price * (1.0 + kappa_p * gap))

    def sell(self, demand):
        sold = min(demand, self.inventory)
        self.sales = sold
        self.inventory -= sold
        return sold

    def compute_revenue(self):
        return self.price * self.sales

    def compute_costs(self):
        interest = 0.05 * self.debt
        wage_bill = self.wage * self.num_employees()
        return wage_bill + interest

    def compute_profit(self):
        rev = self.compute_revenue()
        costs = self.compute_costs()
        self.profit = rev - costs
        return self.profit

    def request_loan(self, needed_amount):
        return max(0.0, needed_amount)

    # --- 新規: 採用・解雇メソッド ---
    def hire_workers(self, candidate_households, n_hires):
        """candidate_households: list of Household objects (失業者候補)
           n_hires: int"""
        if n_hires <= 0 or len(candidate_households) == 0:
            return 0
        n = min(n_hires, len(candidate_households))
        hires = random.sample(candidate_households, n)
        for h in hires:
            h.update_employment(True, firm_id=self.id, wage=self.wage)
            self.employees.append(h)
        return n

    def layoff(self, n_fire):
        """n_fire: number of workers to fire"""
        n_fire = min(n_fire, self.num_employees())
        if n_fire <= 0:
            return 0
        fired = random.sample(self.employees, n_fire)
        for h in fired:
            h.update_employment(False, firm_id=None, wage=0.0)
        self.employees = [e for e in self.employees if e not in fired]
        return n_fire


class Bank:
    def __init__(self, bid, capital, params):
        self.id = bid
        self.capital = float(capital)  # bank own funds
        self.loans = {}  # firm_id -> outstanding loan
        self.params = params
        self.total_deposits = 0.0

    def accept_deposit(self, amount):
        self.total_deposits += amount

    def make_loan(self, firm_id, amount):
        # Check simplistic prudential rule: maintain CRT0 on total assets (loans)
        current_loans = sum(self.loans.values())
        target_crt = self.params['CRT0']
        # allow loan if capital / (capital + new_loans) >= CRT0 roughly
        if (self.capital) / (self.capital + current_loans + amount + 1e-9) >= target_crt:
            self.loans[firm_id] = self.loans.get(firm_id, 0.0) + amount
            # bank extends loan -> no immediate capital change in this simple model,
            # but we track loans; use interest payments later to increase capital
            return amount
        else:
            # partial or zero loan
            # compute max allowed by CRT:
            max_allowed = max(0.0, (self.capital / target_crt) - current_loans - 1e-9)
            loan = min(amount, max_allowed)
            if loan > 0:
                self.loans[firm_id] = self.loans.get(firm_id, 0.0) + loan
            return loan

    def receive_repayment(self, firm_id, repayment):
        outstanding = self.loans.get(firm_id, 0.0)
        repay = min(outstanding, repayment)
        self.loans[firm_id] = outstanding - repay
        # bank receives repayment increases capital (simplified)
        self.capital += repay
        return repay

    def collect_interest(self):
        # interest income from loans increases capital
        income = 0.0
        for fid, outstanding in list(self.loans.items()):
            interest = 0.05 * outstanding
            income += interest
            # interest is paid by firms in the model step when possible (handled elsewhere)
            self.capital += interest
        return income

    def total_outstanding_loans(self):
        return sum(self.loans.values())
def labor_market_matching(unemployed, vacancies, params):
    """労働市場のマッチング関数 (Cobb-Douglas型)"""
    xi = params["xi_match"]
    alpha = params["alpha_match"]
    matches = xi * (vacancies ** alpha) * (unemployed ** (1 - alpha))
    return int(min(matches, unemployed, vacancies))


# --- Model (Scheduler & loop) ---

class CaianiMinimalModel:
    def __init__(self, params):
        self.params = params
        self.households = []
        self.firms = []
        self.banks = []
        self.T = params['T']

        # initializes
        self._init_agents()
        # history
        self.GDP_series = []
        self.Unemployment_series = []
        self.Credit_series = []
        self.Price_series = []
        self.Wage_series = []

    def _init_agents(self):
        p = self.params
        # households
        for i in range(p['N_households']):
            h = Household(i, p['h0_deposit'], p)
            self.households.append(h)

        # firms (consumption firms) — employees start empty
        for j in range(p['N_firms_c']):
            f = FirmC(j, p['f0_capital'], p['f0_debt'], p)
            f.employees = []   # ensure list
            self.firms.append(f)

        # banks
        for bidx in range(p['N_banks']):
            bank = Bank(bidx, p['b0_capital'], p)
            self.banks.append(bank)

        # initial wage baseline
        self.current_wage = p['wn0']

        # --- initial employment assignment ---
        Nc0 = int(p.get('Nc0', int(0.5 * p['N_households'])))
        # sample households to be employed
        employed_ids = random.sample(range(p['N_households']), Nc0)
        for idx, hid in enumerate(employed_ids):
            f = self.firms[idx % len(self.firms)]
            h = self.households[hid]
            f.employees.append(h)
            h.update_employment(True, firm_id=f.id, wage=self.current_wage)

    def step(self, t):
        p = self.params

        # --- 1) wage and unemployment determinations (use households/employee lists)
        total_employed = sum([f.num_employees() for f in self.firms])
        labor_force = p['N_households']   # 全家計を労働力と仮定
        unemployment_rate = max(0.0, 1.0 - (total_employed / max(1, labor_force)))

        # wage adjustment
        kappa_w = 0.05
        self.current_wage = max(0.1, self.current_wage * (1.0 + kappa_w * (p['u_target'] - unemployment_rate)))

        # --- 2) households receive income based on employment and decide consumption
        total_consumption_demand = 0.0
        unemp_benefit = p.get('omega_dole', 0.4) * self.current_wage
        # pay employed households a wage, unemployed get benefit
        for h in self.households:
            if getattr(h, 'employed', False):
                h.receive_income(self.current_wage)
            else:
                h.receive_income(unemp_benefit)
            total_consumption_demand += h.decide_consumption()

        # --- 3) firms produce (use actual employee counts)
        prices = np.array([f.price for f in self.firms])
        # Handle potential NaN or inf values in prices
        prices[np.isnan(prices)] = np.nanmean(prices) if not np.isnan(np.nanmean(prices)) else 1.0 # Replace NaN with mean or 1
        prices[np.isinf(prices)] = np.nanmax(prices) if not np.isinf(np.nanmax(prices)) else 1000.0 # Replace inf with max or a large value

        inv_prices = -prices
        # Handle potential inf values before exp
        inv_prices[np.isinf(inv_prices)] = -1e10  # Replace inf with a large negative number

        exp_scores = np.exp(inv_prices - np.max(inv_prices))

        # Handle potential NaN values after exp
        exp_scores[np.isnan(exp_scores)] = 0.0

        firm_shares = exp_scores / (np.sum(exp_scores) + 1e-12)
        avg_price = np.mean(prices)
        quantity_demand_total = total_consumption_demand / max(1e-6, avg_price)

        for idx, f in enumerate(self.firms):
            firm_demand_qty = firm_shares[idx] * quantity_demand_total
            # Ensure firm_demand_qty is not NaN or inf before converting to int
            if not np.isnan(firm_demand_qty) and not np.isinf(firm_demand_qty):
                # Ensure productivity is not NaN or zero before division
                prod = f.productivity
                if np.isnan(prod) or prod <= 0:
                    prod = 1e-6 # Use a small positive value if productivity is problematic
                labor_required = int(np.ceil(firm_demand_qty / prod))
            else:
                labor_required = 0 # or some other reasonable default

            labor_input = min(labor_required, f.num_employees())
            f.produce(labor_input)

        # --- 4) sales & payments (use num_employees where needed)
        total_sales_value = 0.0
        total_profits = 0.0
        profit_rates = []

        for f in self.firms:
            idx = f.id
            monetary_demand = firm_shares[idx] * total_consumption_demand
            qty_demand = monetary_demand / max(1e-6, f.price)
            sold_qty = f.sell(qty_demand)
            revenue = f.price * sold_qty
            wage_bill = f.num_employees() * self.current_wage
            interest = 0.05 * f.debt
            f.sales = sold_qty
            f.compute_profit()
            profit = f.profit
            total_sales_value += revenue
            total_profits += profit

            # firms attempt to pay wages from revenue; if insufficient, request loan
            funds_needed = max(0.0, wage_bill + interest - revenue)
            if funds_needed > 0:
                bank = random.choice(self.banks)
                loan = bank.make_loan(f.id, funds_needed)
                f.debt += loan

            repay_interest = min(interest, revenue * 0.2)
            if repay_interest > 0:
                for bank in self.banks:
                    if bank.loans.get(f.id, 0.0) > 0:
                        rep = bank.receive_repayment(f.id, repay_interest)
                        f.debt -= rep
                        break

            denom = max(1e-6, revenue)
            profit_rates.append((revenue - (wage_bill + interest)) / denom if denom > 0 else 0.0)

        if len(profit_rates) > 0:
            avg_profit_rate = np.mean(profit_rates)
        else:
            avg_profit_rate = 0.0

        for f in self.firms:
            f.set_price(avg_profit_rate)

        for bank in self.banks:
            bank.collect_interest()

        # --- 5) aggregate bookkeeping and record series
        GDP = total_sales_value
        self.GDP_series.append(GDP)
        # compute unemployment by households (most robust)
        u = 1.0 - (sum(1 for h in self.households if getattr(h, 'employed', False)) / max(1, len(self.households)))
        self.Unemployment_series.append(u)
        total_credit = sum([b.total_outstanding_loans() for b in self.banks])
        self.Credit_series.append(total_credit)
        self.Price_series.append(avg_price)
        self.Wage_series.append(self.current_wage)




    def run(self):
        for t in range(self.T):
            self.step(t)
            # simple progress feedback (every 10%)
            if (t+1) % max(1, int(self.T/10)) == 0:
                print(f"Period {t+1}/{self.T} simulated.")

# --- Execute model ---
if __name__ == '__main__':
    model = CaianiMinimalModel(params)
    model.run()

    # --- Plotting results ---
    periods = range(len(model.GDP_series))

    plt.figure(figsize=(12, 10))

    plt.subplot(4, 1, 1)
    plt.plot(periods, model.GDP_series, label='GDP')
    plt.title('Aggregate Output (GDP)')
    plt.ylabel('GDP (monetary)')
    plt.grid(True)
    plt.legend()

    plt.subplot(4, 1, 2)
    plt.plot(periods, model.Unemployment_series, label='Unemployment Rate')
    plt.title('Unemployment Rate')
    plt.ylabel('Rate')
    plt.grid(True)
    plt.legend()

    plt.subplot(4, 1, 3)
    plt.plot(periods, model.Credit_series, label='Total Bank Credit')
    plt.title('Aggregate Credit')
    plt.ylabel('Credit')
    plt.grid(True)
    plt.legend()

    plt.subplot(4, 1, 4)
    plt.plot(periods, model.Price_series, label='Avg Price')
    plt.plot(periods, model.Wage_series, label='Wage', linestyle='--')
    plt.title('Prices & Wages')
    plt.ylabel('Level')
    plt.xlabel('Time')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python


import numpy as np
import random
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time

# reproducibility
SEED = 12345
random.seed(SEED)
np.random.seed(SEED)

# --- Baseline params (Caiani - from provided table) ---
params = {
    # scale (試験的に小さくして試すならここを変更)
    'N_households': 8000,
    'N_firms_c': 100,
    'N_firms_k': 20,
    'N_banks': 10,
    'Nc0': int(0.5 * 8000),

    # time
    'T': 200,

    # macro
    'gSS': 0.0075,

    # tech / production
    'muN': 2.0,
    'mu_k': 1.0,
    'lk': 6.4,

    # market matching/contact
    'chi_c': 5,
    'chi_k': 5,
    'chi_d': 3,
    'chi_l': 3,
    'chi_n': 10,

    # selection intensities
    'eps_d': 2.00687,
    'eps_l': 2.00687,
    'eps_c': 1.50515,
    'eps_k': 1.50515,

    # inventories, expectations
    'nu': 0.1,
    'lambda_adapt': 0.25,
    'theta_turnover': 0.05,

    # price/wage noise
    'mu_FN': 1.0,
    'sigma2_FN': 0.0094,

    # taxation
    'tau_pi': 0.18,
    'tau_i': 0.18,

    # lending / capital
    'eta': 20,
    'kappa': 20,
    'r_target': 0.04345,
    'u_target': 0.8,

    # investment weights
    'gamma1': 0.01,
    'gamma2': 0.02,

    # buffers, payout rates
    'sigma_buffer': 1.0,
    'rho_c': 0.9,
    'rho_k': 0.9,
    'rho_b': 0.6,

    # asset / bond
    'i_b0': 0.0025,
    'p_b': 1.0,

    # bank regulation targets
    'CRT0': 0.17996,
    'LRT0': 0.258026342,

    # bank risk aversion
    's_c': 3.92245,
    's_k': 21.51335,

    # bankruptcy haircut
    'haircut': 0.5,

    # wage / unemployment
    'wn0': 5.0,
    'omega_dole': 0.4,
    'v_unemp': 0.08,

    # consumption propensities
    'alpha1': 0.38581,
    'alpha2': 0.25,

    # initial stocks
    'h0_deposit': 50.0,
    'f0_capital': 500.0,
    'f0_debt': 50.0,
    'b0_capital': 50000.0,

    # missing / added
    'markup_c0': 0.318857,
    "xi_match": 0.5,
    "alpha_match": 0.6,
    "eta_w": 0.1,
    "w_ref": 5.0,
}

# --- Agent classes ---

class Household:
    def __init__(self, hid, deposit, params):
        self.id = hid
        self.deposit = float(deposit)
        self.income = 0.0
        self.wealth = float(deposit)
        self.consumption = 0.0
        self.params = params
        self.employed = False
        self.firm_id = None
        self.wage = 0.0

    def update_employment(self, employed, firm_id=None, wage=0.0):
        self.employed = bool(employed)
        self.firm_id = firm_id if employed else None
        self.wage = float(wage) if employed else 0.0

    def receive_income(self, wage_income):
        self.income = float(wage_income)
        self.deposit += self.income
        self.wealth = self.deposit

    def decide_consumption(self):
        a1 = self.params['alpha1']
        a2 = self.params['alpha2']
        self.consumption = a1 * self.income + a2 * self.wealth
        if self.consumption > self.deposit:
            self.consumption = self.deposit
        return self.consumption

    def pay(self, amount):
        paid = min(amount, self.deposit)
        self.deposit -= paid
        self.wealth = self.deposit
        return paid

class FirmC:
    def __init__(self, fid, capital, debt, params):
        self.id = fid
        self.capital = float(capital)
        self.debt = float(debt)
        self.cash = float(capital)   # 追加: 現金アカウント
        self.inventory = 0.0
        self.price = 1.0 + params['markup_c0']
        self.wage = params['w_ref']
        self.productivity = 1.0
        self.employees = []
        self.output = 0.0
        self.sales = 0.0
        self.profit = 0.0
        self.params = params
        self.past_demand = []
        self.inv_target_rate = params['nu']
        self.vacancies = 0
        self._labor_required = 0

    def num_employees(self):
        return len(self.employees)

    def produce(self, labor_input=None):
        L = self.num_employees() if labor_input is None else labor_input
        A = self.productivity
        self.output = A * L
        self.inventory += self.output

    def set_price(self, aggregate_avg_profit):
        kappa_p = 0.02
        if self.sales > 0:
            pi_i = (self.price * self.sales - self.wage * self.num_employees() - 0.05 * self.debt) / (self.price * self.sales)
        else:
            pi_i = 0.0
        gap = pi_i - aggregate_avg_profit
        self.price = max(0.01, self.price * (1.0 + kappa_p * gap))

    def sell(self, demand):
        sold = min(demand, self.inventory)
        self.sales = sold
        self.inventory -= sold
        return sold

    def compute_revenue(self):
        return self.price * self.sales

    def compute_costs(self):
        interest = 0.05 * self.debt
        wage_bill = self.wage * self.num_employees()
        return wage_bill + interest

    def compute_profit(self):
        rev = self.compute_revenue()
        costs = self.compute_costs()
        self.profit = rev - costs
        return self.profit

    def request_loan(self, needed_amount):
        return max(0.0, needed_amount)

    def hire_workers(self, candidate_households, n_hires):
        if n_hires <= 0 or len(candidate_households) == 0:
            return 0
        n = min(n_hires, len(candidate_households))
        hires = random.sample(candidate_households, n)
        for h in hires:
            h.update_employment(True, firm_id=self.id, wage=self.wage)
            self.employees.append(h)
        return n

    def layoff(self, n_fire):
        n_fire = min(n_fire, self.num_employees())
        if n_fire <= 0:
            return 0
        fired = random.sample(self.employees, n_fire)
        for h in fired:
            h.update_employment(False, firm_id=None, wage=0.0)
        self.employees = [e for e in self.employees if e not in fired]
        return n_fire

class Bank:
    def __init__(self, bid, capital, params):
        self.id = bid
        self.capital = float(capital)
        self.loans = {}  # firm_id -> outstanding loan
        self.params = params
        self.total_deposits = 0.0

    def accept_deposit(self, amount):
        self.total_deposits += amount

    def make_loan(self, firm_id, amount):
        current_loans = sum(self.loans.values())
        target_crt = self.params['CRT0']
        # very simple prudential check
        if (self.capital) / (self.capital + current_loans + amount + 1e-9) >= target_crt:
            self.loans[firm_id] = self.loans.get(firm_id, 0.0) + amount
            return amount
        else:
            max_allowed = max(0.0, (self.capital / target_crt) - current_loans - 1e-9)
            loan = min(amount, max_allowed)
            if loan > 0:
                self.loans[firm_id] = self.loans.get(firm_id, 0.0) + loan
            return loan

    def receive_repayment(self, firm_id, repayment):
        outstanding = self.loans.get(firm_id, 0.0)
        repay = min(outstanding, repayment)
        self.loans[firm_id] = outstanding - repay
        self.capital += repay
        return repay

    def collect_interest(self):
        income = 0.0
        for fid, outstanding in list(self.loans.items()):
            interest = 0.05 * outstanding
            income += interest
            self.capital += interest
        return income

    def total_outstanding_loans(self):
        return sum(self.loans.values())

def labor_market_matching(unemployed, vacancies, params):
    xi = params["xi_match"]
    alpha = params["alpha_match"]
    matches = xi * (vacancies ** alpha) * (unemployed ** (1 - alpha))
    return int(min(matches, unemployed, vacancies))

# --- Model ---

class CaianiMinimalModel:
    def __init__(self, params):
        self.params = params
        self.households = []
        self.firms = []
        self.banks = []
        self.T = params['T']

        self._init_agents()
        self.GDP_series = []
        self.Unemployment_series = []
        self.Credit_series = []
        self.Price_series = []
        self.Wage_series = []
        self.Balance_check = []  # store (household_deposits, firm_cash_sum, bank_capital_sum, total_loans)

    def _init_agents(self):
        p = self.params
        for i in range(p['N_households']):
            h = Household(i, p['h0_deposit'], p)
            self.households.append(h)

        for j in range(p['N_firms_c']):
            f = FirmC(j, p['f0_capital'], p['f0_debt'], p)
            f.employees = []
            self.firms.append(f)

        for bidx in range(p['N_banks']):
            bank = Bank(bidx, p['b0_capital'], p)
            self.banks.append(bank)

        self.current_wage = p['wn0']

        Nc0 = int(p.get('Nc0', int(0.5 * p['N_households'])))
        employed_ids = random.sample(range(p['N_households']), Nc0)
        for idx, hid in enumerate(employed_ids):
            f = self.firms[idx % len(self.firms)]
            h = self.households[hid]
            f.employees.append(h)
            h.update_employment(True, firm_id=f.id, wage=self.current_wage)

    def step(self, t, verbose=False):
        p = self.params

        # 1) unemployment & wage update
        total_employed = sum([f.num_employees() for f in self.firms])
        labor_force = p['N_households']
        unemployment_rate = max(0.0, 1.0 - (total_employed / max(1, labor_force)))
        kappa_w = 0.05
        self.current_wage = max(0.1, self.current_wage * (1.0 + kappa_w * (p['u_target'] - unemployment_rate)))
        unemp_benefit = p.get('omega_dole', 0.4) * self.current_wage

        # 2) Firms pay wages from cash (use previous-period cash), if insufficient borrow
        #    pay wages BEFORE households decide consumption (income affects consumption)
        for f in self.firms:
            wage_bill = f.num_employees() * self.current_wage
            if wage_bill > 0:
                if f.cash >= wage_bill:
                    f.cash -= wage_bill
                    # transfer to households (do per-worker)
                    for worker in f.employees:
                        worker.receive_income(self.current_wage)
                else:
                    needed = wage_bill - f.cash
                    # borrow from a random bank (simplification)
                    bank = random.choice(self.banks)
                    loan = bank.make_loan(f.id, needed)
                    f.debt += loan
                    f.cash += loan
                    # if still short, pay what you can (we allow negative cash if loan insufficient)
                    pay_now = min(f.cash, wage_bill)
                    f.cash -= pay_now
                    # distribute pay_now equally to employees
                    if f.num_employees() > 0:
                        per_worker = pay_now / f.num_employees()
                    else:
                        per_worker = 0.0
                    for worker in f.employees:
                        worker.receive_income(per_worker)
                    # if loan was less than needed, remaining unpaid wages are 0 (could be sticky)
            # if no employees, nothing to do

        # for unemployed households, give unemployment benefit
        for h in self.households:
            if not h.employed:
                h.receive_income(unemp_benefit)

        # 3) Households decide consumption (desired)
        total_consumption_demand = 0.0
        for h in self.households:
            total_consumption_demand += h.decide_consumption()

        # 4) Firms produce based on employees and expected labor requirement (we base on price shares)
        prices = np.array([f.price for f in self.firms])
        prices[np.isnan(prices)] = np.nanmean(prices) if not np.isnan(np.nanmean(prices)) else 1.0
        prices[np.isinf(prices)] = np.nanmax(prices) if not np.isinf(np.nanmax(prices)) else 1000.0
        inv_prices = -prices
        inv_prices[np.isinf(inv_prices)] = -1e10
        exp_scores = np.exp(inv_prices - np.max(inv_prices))
        exp_scores[np.isnan(exp_scores)] = 0.0
        firm_shares = exp_scores / (np.sum(exp_scores) + 1e-12)
        avg_price = np.mean(prices)

        # compute desired quantity demand per firm
        quantity_demand_total = total_consumption_demand / max(1e-6, avg_price)
        total_expected_qty = 0.0
        for idx, f in enumerate(self.firms):
            firm_demand_qty = firm_shares[idx] * quantity_demand_total
            prod = f.productivity
            if np.isnan(prod) or prod <= 0:
                prod = 1e-6
            labor_required = int(np.ceil(firm_demand_qty / prod))
            f._labor_required = labor_required
            # produce using current employees (can't instantly hire here)
            labor_input = min(labor_required, f.num_employees())
            f.produce(labor_input)
            total_expected_qty += firm_demand_qty

        # 5) Sales: firms sell up to demand (monetary demand -> qty) and compute revenue
        total_sales_value = 0.0
        total_revenue = 0.0
        for idx, f in enumerate(self.firms):
            monetary_demand = firm_shares[idx] * total_consumption_demand
            qty_demand = monetary_demand / max(1e-6, f.price)
            sold_qty = f.sell(qty_demand)
            revenue = f.price * sold_qty
            f.sales = sold_qty
            f.compute_profit()
            total_sales_value += revenue
            total_revenue += revenue

        # 6) Households actually pay: we conserve money by withdrawing from households proportionally
        #    to their desired consumption, scaled by (total_revenue / total_consumption_demand)
        if total_consumption_demand <= 0:
            pay_frac = 0.0
        else:
            pay_frac = min(1.0, total_revenue / total_consumption_demand)

        total_collected = 0.0
        for h in self.households:
            intended = h.consumption
            to_pay = intended * pay_frac
            paid = h.pay(to_pay)
            total_collected += paid

        # total_collected should be approximately total_revenue (floating rounding)
        # Allocate collected money to firms' cash based on each firm's revenue share
        if total_revenue > 0:
            for f in self.firms:
                rev = f.compute_revenue()
                alloc = rev  # revenue is what firm sold and should receive
                f.cash += alloc
        else:
            # no revenue: nothing allocated
            pass

        # 7) Firms attempt to pay remaining obligations: interest, wages (if unpaid), etc.
        for f in self.firms:
            # interest
            interest = 0.05 * f.debt
            wage_bill = f.num_employees() * self.current_wage
            # try to pay interest first out of cash
            repay_interest = min(interest, f.cash)
            if repay_interest > 0:
                # repay to any bank that has loan for this firm
                remaining = repay_interest
                for bank in self.banks:
                    outstanding = bank.loans.get(f.id, 0.0)
                    if outstanding > 0 and remaining > 0:
                        rep = min(outstanding, remaining)
                        bank.receive_repayment(f.id, rep)
                        f.debt -= rep
                        f.cash -= rep
                        remaining -= rep
                    if remaining <= 0:
                        break

            # if wage needs to be (re)paid this period (we did pre-pay above from previous cash),
            # any shortfalls could be addressed now:
            wage_needed = wage_bill - 0.0  # wage already distributed earlier in this step
            if wage_needed > 0:
                # should rarely happen given earlier payments, but handle generically
                if f.cash < wage_needed:
                    needed = wage_needed - f.cash
                    bank = random.choice(self.banks)
                    loan = bank.make_loan(f.id, needed)
                    f.debt += loan
                    f.cash += loan
                pay_now = min(f.cash, wage_needed)
                f.cash -= pay_now
                # distribute pay_now equally
                if f.num_employees() > 0:
                    per_worker = pay_now / f.num_employees()
                else:
                    per_worker = 0.0
                for worker in f.employees:
                    worker.receive_income(per_worker)

            # if cash still negative (loan limits), allow negative for now or set to zero
            # (we keep it simple)
            f.compute_profit()

        # 8) Labor market matching -> hires & layoffs
        unemployed_list = [h for h in self.households if not h.employed]
        total_unemp = len(unemployed_list)
        vacancies_list = [max(0, f._labor_required - f.num_employees()) for f in self.firms]
        total_vacancies = sum(vacancies_list)
        matches = 0
        if total_vacancies > 0 and total_unemp > 0:
            matches = labor_market_matching(total_unemp, total_vacancies, p)

        if matches > 0:
            random.shuffle(unemployed_list)
            assigned = 0
            for f, vac in zip(self.firms, vacancies_list):
                if vac <= 0:
                    continue
                # proportionally allocate matches to this firm's vacancies
                hires = int(round(matches * vac / max(1, total_vacancies)))
                hires = min(hires, len(unemployed_list) - assigned)
                if hires <= 0:
                    continue
                for i in range(hires):
                    h = unemployed_list[assigned]
                    h.update_employment(True, firm_id=f.id, wage=self.current_wage)
                    f.employees.append(h)
                    assigned += 1
                if assigned >= matches:
                    break

        # layoffs: if f.num_employees() > _labor_required -> fire some
        for f in self.firms:
            over = f.num_employees() - f._labor_required
            if over > 0:
                f.layoff(over)

        # 9) Banks collect interest (adds to capital)
        for bank in self.banks:
            bank.collect_interest()

        # 10) Record aggregates and simple balance check
        GDP = total_sales_value
        self.GDP_series.append(GDP)
        u = 1.0 - (sum(1 for h in self.households if getattr(h, 'employed', False)) / max(1, len(self.households)))
        self.Unemployment_series.append(u)
        total_credit = sum([b.total_outstanding_loans() for b in self.banks])
        self.Credit_series.append(total_credit)
        self.Price_series.append(avg_price)
        self.Wage_series.append(self.current_wage)

        # balance check
        hh_deposits = sum(h.deposit for h in self.households)
        firms_cash = sum(f.cash for f in self.firms)
        banks_capital = sum(b.capital for b in self.banks)
        total_loans = total_credit
        self.Balance_check.append((hh_deposits, firms_cash, banks_capital, total_loans))

        if verbose and (t % max(1, int(self.T/10)) == 0):
            print(f"t={t}: GDP={GDP:.2f}, Unemp={u:.3f}, Credit={total_credit:.2f}")

    def run(self, show_progress=True):
        for t in range(self.T):
            self.step(t, verbose=False)
            if show_progress:
                if (t+1) % max(1, int(self.T/20)) == 0 or t==0 or t==self.T-1:
                    clear_output(wait=True)
                    print(f"Simulating period {t+1}/{self.T}")
                    # show brief diagnostics
                    hh_deposits, firms_cash, banks_capital, total_loans = self.Balance_check[-1]
                    print(f"Household deposits: {hh_deposits:.2f}, Firms cash: {firms_cash:.2f}, Banks capital: {banks_capital:.2f}, Total loans: {total_loans:.2f}")
                    time.sleep(0.1)

# --- Execute model ---
if __name__ == '__main__':
    model = CaianiMinimalModel(params)
    # Optional: for quicker debug, reduce T in params or override here, e.g. model.T = 50
    model.run(show_progress=True)

    # --- Plot results ---
    periods = range(len(model.GDP_series))
    plt.figure(figsize=(12, 10))

    plt.subplot(4, 1, 1)
    plt.plot(periods, model.GDP_series, label='GDP')
    plt.title('Aggregate Output (GDP)')
    plt.ylabel('GDP (monetary)')
    plt.grid(True)
    plt.legend()

    plt.subplot(4, 1, 2)
    plt.plot(periods, model.Unemployment_series, label='Unemployment Rate')
    plt.title('Unemployment Rate')
    plt.ylabel('Rate')
    plt.grid(True)
    plt.legend()

    plt.subplot(4, 1, 3)
    plt.plot(periods, model.Credit_series, label='Total Bank Credit')
    plt.title('Aggregate Credit')
    plt.ylabel('Credit')
    plt.grid(True)
    plt.legend()

    plt.subplot(4, 1, 4)
    plt.plot(periods, model.Price_series, label='Avg Price')
    plt.plot(periods, model.Wage_series, label='Wage', linestyle='--')
    plt.title('Prices & Wages')
    plt.ylabel('Level')
    plt.xlabel('Time')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()
```
